```python
ott
```

```python
ott['sale'] = [185900000, 973000000, 217000000, 212400000, 415400000, 180200000, 377000000, 220100000, 631700000,230100000,663000000,292400000,773300000,273500000,592000000,344900000,823300000,248000000,430000000,365300000]
ott
```

```python
movie = movie[movie['year'].between(2020,2023)]
ott = ott[ott['year'].between(2020,2023)]

# 연도별 평균 매출 계산
movie_avg_sales = movie.groupby("year")["sale"].mean().reset_index()
ott_avg_sales = ott.groupby("year")["sale"].mean().reset_index()
```

```python

plt.figure(figsize=(12, 6))

# 영화 평균 매출
plt.bar(movie_avg_sales["year"], movie_avg_sales["sale"],color='skyblue')
# OTT 평균 매출
plt.plot(ott_avg_sales["year"], ott_avg_sales["sale"], marker="o", color="orange", label="OTT")

plt.title("영화와 OTT의 연도별 평균 매출 비교", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Sales", fontsize=12)
plt.xticks(movie_avg_sales["year"], rotation=45)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.savefig("ott_and_movie_sale.png")
plt.show()
```

### 2-3. 장르별 추이 

```python
netflix
```

```python
tving
```

```python
ott_ranking = pd.concat([netflix,tving], ignore_index = True)
ott_ranking
```

```python
ott_ranking = ott_ranking.drop_duplicates(subset="name")
ott_ranking 
```

```python
ott_ranking['genre'] = ott_ranking['genre'].replace({'로맨스': '멜로/로맨스','멜로/멜로/로맨스':'멜로/로맨스', '어드벤처': '어드벤처(모험)'}, regex=True)
ott_ranking
```

```python
ott_ranking['genre'] = ott_ranking['genre'].replace({'가족': '기타', '서스펜스': '기타'}, regex=True)
ott_ranking
```

```python
ott_ranking = ott_ranking[~ott_ranking['genre'].str.contains('예능')]
ott_ranking
```

```python

# 장르를 쉼표로 분리하여 리스트로 변환
genre_split = ott_ranking['genre'].str.split(',').explode().str.strip()

# 각 장르별 횟수 세기
genre_counts = genre_split.value_counts()

# 결과 출력
print(genre_counts)
```

```python
genre_counts
```

```python
genre = pd.DataFrame(genre_counts)
genre
```

```python
genre = {
    "genre": [
        "드라마", "기타", "액션", "미스터리", 
        "어드벤처(모험)", "판타지", 
        "스릴러", "범죄", "멜로/멜로/멜로/로맨스", 
        "코미디", "애니메이션", "멜로/멜로/로맨스", 
        "공포(호러)", "SF", "전쟁"
    ],
    "count": [12, 8, 8, 8, 8, 7, 6, 4, 3, 3, 3, 3, 2, 1, 1]
}
```

```python
df = pd.DataFrame(genre)
# 장르 정제
df['genre'] = df['genre'].str.replace(r'멜로/멜로/멜로/로맨스', '멜로/로맨스')  # 중복 멜로/로맨스 정리
df['genre'] = df['genre'].str.replace(r'멜로/멜로/로맨스', '멜로/로맨스')  # 추가 정리

# 중복 장르 합산
df = df.groupby('genre', as_index=False).sum()

# 결과 출력
df
```

```python

plt.figure(figsize=(10, 6))
plt.bar(df["genre"], df["count"], color='skyblue')
plt.title("top 20 장르", fontsize=16)
plt.xlabel("장르", fontsize=12)
plt.ylabel("작품 수", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("ott_genre.png")
plt.show()
```
